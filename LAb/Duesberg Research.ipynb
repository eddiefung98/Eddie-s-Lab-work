```python
def convert_df(file):
    df = pd.read_csv(file).drop([0,1,2])
    df.columns = df.iloc[0]
    df.drop([3], inplace = True)
    df.rename(columns={ df.columns[0]: "Chromosome #" }, inplace = True)
    df.dropna(axis = 1, how = 'all', inplace = True)
    df = df.iloc[:, :-2]
    df.set_index('Chromosome #', inplace = True)
    df.dropna(axis = 0, how = 'all', inplace = True)
    df.fillna(0, inplace = True)
    return df
```

```python
def convert_df_weird_format(file):
    df = pd.read_csv(file)
    df.columns = df.iloc[1]
    df.drop([3], inplace = True)
    df.rename(columns={ df.columns[0]: "Chromosome #" }, inplace = True)
    df.dropna(axis = 1, how = 'all', inplace = True)
    df = df.iloc[:, :-2]
    df.dropna(axis = 0, how = 'any', inplace = True)
    df.set_index('Chromosome #', inplace = True)
    df.dropna(axis = 0, how = 'all', inplace = True)
    df.fillna(0, inplace = True)
    df.drop(df.index[0], inplace = True)
    return df
```

```python
def sample(df):
    sample_df = pd.DataFrame()
    n = 0
    for i in np.arange(5000):
        num = np.random.choice(len(df.columns))
        col = df.iloc[:, num]
        sample_df['K' + str(n)] = col
        n+= 1
    return sample_df
```

```python
def pca(df, name):
    #if type(df) == pd.core.frame.DataFrame:
    df = df.apply(pd.to_numeric)
    matrix = df.values
    #if type(df) == numpy.ndarray:
        #matrix = df
    n = matrix.shape[1]
    X = (matrix.T - np.mean(matrix, axis=1)) / np.sqrt(n)
    u, s, vt = np.linalg.svd(X, full_matrices=False)
    t = X @ vt.T
    total_var = sum (s ** 2)
    proportion1, proportion2 = s[0] ** 2 / total_var, (s[1] ** 2 + s[0] ** 2) / total_var
    print(name, proportion1, proportion2)
    sns.scatterplot(x=t[:, 0], y=t[:, 1], label = name)
    plt.xlabel('PC1')
    plt.ylabel('PC2')
```

```python
def pcakde(df, name):
    #if type(df) == pd.core.frame.DataFrame:
    df = df.apply(pd.to_numeric)
    matrix = df.values
    #if type(df) == numpy.ndarray:
        #matrix = df
    n = matrix.shape[0]
    X = (matrix - np.mean(matrix, axis=0)) / np.sqrt(n)
    u, s, vt = np.linalg.svd(X, full_matrices=False)
    t = X @ vt.T
    total_var = sum (s ** 2)
    proportion1, proportion2 = s[0] ** 2 / total_var, (s[1] ** 2 + s[0] ** 2) / total_var
    print(name, proportion1, proportion2)
    sns.kdeplot(x=t[:, 0], y=t[:, 1], label = name)
    plt.xlabel('PC1')
    plt.ylabel('PC2')
```
